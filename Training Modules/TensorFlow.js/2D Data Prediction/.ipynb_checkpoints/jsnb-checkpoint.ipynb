```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import tensorflow as tf
import tensorflowjs as tfjs
from tensorflow import keras
```

```python
df = pd.read_json("data.json")
df = df.dropna()
```

```python
sns.scatterplot(
    data=df, x=df["Horsepower"], y=df["Miles_per_Gallon"], 
    size=df["Cylinders"], hue=df["Displacement"]
)
```

```python
inputs = np.array(df["Miles_per_Gallon"], dtype=float)
labels = np.array(df["Horsepower"], dtype=float)
```

```python
input_max = inputs.max()
input_min = inputs.min()
label_max = labels.max()
label_min = labels.min()

# EXPECT A S#$% TON OF NANS IF YOU DON'T NORMALIZE!
normalized_inputs = (inputs - input_min) / (input_max - input_min)
normalized_labels = (labels - label_min) / (label_max - label_min)
```

```python
# Define a layer
layer_0 = keras.layers.Dense(units=1, input_shape=[1], use_bias=True)
# Define a model with a list of layers
model = tf.keras.Sequential([layer_0]) 
model.compile(optimizer='sgd', loss='mean_squared_error', metrics="mse")
model.fit(x=normalized_inputs, y=normalized_labels, epochs=50, verbose=2, validation_split=0.3, shuffle=False)
```
